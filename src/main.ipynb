```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df_1 = pd.read_csv(f"../data/H2AFZ.parathyroid_adenoma.ENCFF042UAX.hg19.bed", header=None, sep="\t")
```

```python
df_2 = pd.read_csv(f"../data/H2AFZ.parathyroid_adenoma.ENCFF452OXY.hg19.bed", header=None, sep="\t")
```

```python
lens_1 = df_1[2] - df_1[1]
```

```python
lens_2 = df_2[2] - df_2[1]
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(1, 2, 1)
plt.hist(lens_1, color="green", bins=50)
plt.xlabel("Length")
plt.ylabel("Count")
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF042UAX.hg19\nNumber of peaks: {len(lens_1)}")

plt.subplot(1, 2, 2)
plt.hist(lens_2, color="green", bins=50)
plt.xlabel("Length")
plt.ylabel("Count")
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF452OXY.hg19\nNumber of peaks: {len(lens_2)}")

plt.savefig("../images/len_hist.png", format="png")

plt.show()
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(1, 2, 1)
plt.boxplot(lens_1)
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF042UAX.hg19 boxplot")

plt.subplot(1, 2, 2)
plt.boxplot(lens_2)
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF452OXY.hg19 boxplot")

plt.savefig("../images/len_boxplot.png", format="png")

plt.show()
```

```python
filter_length = 2000

plt.figure(figsize=(14, 4))

plt.subplot(1, 2, 1)
plt.hist(lens_1[lens_1 < filter_length], color="green", bins=50)
plt.xlabel("Length")
plt.ylabel("Count")
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF042UAX.hg19\nFiltered peaks with lengths greater than {filter_length}\nNumber of peaks: {(lens_1 < filter_length).sum()}")

plt.subplot(1, 2, 2)
plt.hist(lens_2[lens_2 < filter_length], color="green", bins=50)
plt.xlabel("Length")
plt.ylabel("Count")
plt.title(f"H2AFZ.parathyroid_adenoma.ENCFF452OXY.hg19\nFiltered peaks with lengths greater than {filter_length}\nNumber of peaks: {(lens_2 < filter_length).sum()}")

plt.savefig("../images/filtered_len_hist.png", format="png")

plt.show()
```

```python
df_1[lens_1 < filter_length].to_csv("../data/H2AFZ.parathyroid_adenoma.ENCFF042UAX.hg19.filtered.bed", index=None, header=None, sep="\t")
df_2[lens_2 < filter_length].to_csv("../data/H2AFZ.parathyroid_adenoma.ENCFF452OXY.hg19.filtered.bed", index=None, header=None, sep="\t")
```
